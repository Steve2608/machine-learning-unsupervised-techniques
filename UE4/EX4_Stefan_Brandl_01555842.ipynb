# Exercise A

In this exercise we want to discuss certain theoretical basics of ICA in greater detail. An important aspect of ICA is concerned with measuring non-Gaussianity of random variables. Here we discuss one possible approach in this context that uses cumulants of random variables. These objects are in general defined as follows: for a real-valued random variable $X$ we consider the cumulant generating function $K_{X}$ given by $K_{X}(t)=\ln(\text{E}(e^{Xt}))$. The $n$-th order cumulant $\kappa_n(X)$ corresponding to $X$ is then defined as the $n$-th coefficient in the Taylor-series expansion for $K_{X}(t)$ near $0$, i.e. $K_{X}(t)=\sum_{n=1}^\infty \frac{\kappa_n( X) t^n}{n!}$, and thus $\kappa_n(X)=K_{X}^{(n)}(0)$ (the $n$-th derivative of $K_{X}(t)$ evaluated at $0$). We start by showing some important properties of cumulants, that make many calculations (especially regarding sums of independent random variables) easier compared to the use of moments. Let us try to get a better understanding about how these objects look like explicitely and can be computed in practice by deriving relations between moments and cumulants.

<h3 style="color:rgb(208,90,80)">Answer 4.4:</h3>

\begin{align*}
K_4(S_n) &= K_4(\frac{X_1 + \dots + X_n}{\sqrt{n}}) \\
&= (\frac{1}{\sqrt{n}})^n \sum_i^n K_4(X_i) \\
&= \sum_i^n \frac{K_4(X_i)}{\sqrt{n}^n}
\end{align*}

Now let n approach $\infty$:

\begin{align*}
\lim\limits_{n \rightarrow \infty}{\sum_i^n \frac{K_4(X_i)}{\sqrt{n}^n}}
\end{align*}

since: 

\begin{align*}
\lim\limits_{n \rightarrow \infty}{\sqrt{n}^n} \rightarrow \infty
\end{align*}

we now that:

\begin{align*}
\lim\limits_{n \rightarrow \infty}{\sum_i^n \frac{K_4(X_i)}{\sqrt{n}^n}} \rightarrow 0
\end{align*}

Since $\kappa_4(X)=0$ for a standard normally distributed random variable $X \sim \mathcal{N}(0,1)$, this is in perfect accordance with the central limit theorem, which basically tells you, that sums of normalized and rescaled independent random variables approach $X \sim \mathcal{N}(0,1)$. This should give you a first theoretical explanation, why cumulants, and especially the forth cumulant, can be a reasonable concept to measure non-Gaussianity.

To this end, compute the cumulants for a $\lambda$-exponentially distributed variable $X$. Recall, that the density of the exponential distribution is given by
$$
      p_{\lambda}(x) = \begin{cases}\displaystyle
                         \lambda {\rm e}^{-\lambda x} & x \ge 0 \\
			0 & x < 0
                      \end{cases} 
$$
for any $\lambda \geq 0$. As a result you should obtain $\kappa_n(X)=\frac{(n-1)!}{\lambda^n}$.

<h3 style="color:rgb(208,90,80)">Answer 4.5:</h3>

We now comupute the moment generating function of the given distribution: 

\begin{align*}
E_X(e^{tX}) &= \int_{-\infty}^\infty e^{tX} p_\lambda(x) dx \\
&= \underbrace{\int_{-\infty}^\infty}_\text{All values $x < 0$ are $0$} e^{tx} \lambda e^{-\lambda x} dx \\
&= \int_0^\infty e^{tx} \lambda e^{-\lambda x} dx \\
&= \lambda \int_0^\infty e^{tx} e^{-\lambda x} dx \\
&= \lambda \int_0^\infty e^{tx - \lambda x} dx \\
&= \lambda \int_0^\infty e^{(t - \lambda)x} dx \\
&= \lambda \frac{1}{t - \lambda} \big[e^{(t - \lambda)x}\big]^\infty_0 \\
&= \frac{\lambda}{t - \lambda} \big[e^{(t - \lambda)x}\big]^\infty_0 \\
&= \frac{\lambda}{t - \lambda} \big(e^{(t - \lambda) \infty} - e^{(t - \lambda) 0}\big) \\
&= \frac{\lambda}{t - \lambda} \big(e^{(\lambda-t) (-\infty)} - e^0 \big) \\
&= \frac{\lambda}{t - \lambda} \big(\underbrace{e^{(\lambda-t) (-\infty)}}_\text{0 for $t > \lambda$}  - 1 \big) \\
&= \frac{\lambda}{t - \lambda} \cdot (-1) \\
&= \frac{\lambda}{\lambda - t}
\end{align*}

The cumulant generating function is defined as:

\begin{align*}
K_X(t) = \ln(\frac{\lambda}{\lambda - t})
\end{align*}

We now generate some cumulants: 

\begin{align*}
K_1(t) &= K'_X(0) \\
&= \frac{1}{\lambda - t} \xrightarrow{t = 0} \frac{1}{\lambda}
\end{align*}

\begin{align*}
K_2(t) &= K'^{(2)}_X(0) \\
&= \big( \frac{1}{\lambda - t} \big)' \\
&= \frac{1}{(\lambda - t)^2} \xrightarrow{t = 0} \frac{1}{\lambda^2}
\end{align*}

\begin{align*}
K_3(t) &= K'^{(3)}_X(0) \\
&= \big( \frac{1}{(\lambda - t)^2} \big)' \\
&= \frac{2}{(\lambda - t)^3} \xrightarrow{t = 0} \frac{2}{\lambda^3}
\end{align*}

\begin{align*}
K_4(t) &= K'^{(4)}_X(0) \\
&= \big( \frac{2}{(\lambda - t)^3} \big)' \\
&= \frac{6}{(\lambda - t)^4} \xrightarrow{t = 0} \frac{6}{\lambda^4}
\end{align*}

\begin{align*}
K_5(t) &= K'^{(5)}_X(0) \\
&= \big( \frac{6}{(\lambda - t)^4} \big)' \\
&= \frac{24}{(\lambda - t)^5} \xrightarrow{t = 0} \frac{24}{\lambda^5}
\end{align*}

Which generalises to 

\begin{align*}
K_n(t) &= K'^{(n)}_X(0) \\
&= \big( \frac{(n-2)!}{(\lambda - t)^{n-1}} \big)' \\
&= \frac{(n-1)!}{(\lambda - t)^{n}} \xrightarrow{t = 0} \frac{(n-1)!}{\lambda^{n}}
\end{align*}

Let us mention that the previous argument easily carries over to a Laplace-distributed random variable $X$, which is just a shifted and symmetrized version of the exponential distribution. More preciesly, it has a location parameter $\mu \in \mathbb{R}$ and scaling parameter $\lambda>0$ and is defined by the density 
$$
      p_{\mu,\lambda}(x) = \begin{cases}\displaystyle
                         \frac{\lambda}{2} {\rm e}^{-(x-\mu) \lambda} & x \ge \mu \\
			\frac{\lambda}{2} {\rm e}^{-(\mu-x)\lambda} & x < \mu.
                      \end{cases} 
$$
By using exactly the same reasoning as before(for each side around $\mu$), one obtains $\kappa_n(X)=\frac{2(n-1)!}{\lambda^n}$, so the cumulants behave the same way as for the exponential distribution.

The next plot should further strengthen your intuition about how deviation from Gaussianity looks like for the Laplace distributed random variables and how this is affected by the parameter $\lambda$. 

Laplace random variables are a typical example of supergaussian random variables, they 
have a spiky density function with heavy tails, i.e. the density function is relatively large at $0$ and at large values, while relatively small for intermediate ones. 

For these kind of distributions (or other kinds of heavy tailed distributions), cumulants are indeed very useful objects to measure the deviation from normality.

<h3 style="color:rgb(208,90,80)">Answer 4.6:</h3>

```python
from scipy import signal
from sklearn.decomposition import FastICA, PCA

# step 1
interval = np.linspace(-1, 1, 2001)

# step 2
s = np.sin(interval)
c = np.cos(interval)
t = np.tan(interval)
st = signal.sawtooth(interval)

# step 3
original = np.column_stack((s, c, t, st))

# step 4
def noise(mat, scale):
    return mat + np.random.normal(0, scale, mat.shape)

noisy = noise(original, .15)
```

```python
def plot_periodical(matrices, labels, colors, titles=['Sine', 'Cosine', 'Tangent', 'Sawtooth']):
    _, ax = plt.subplots(nrows=2, ncols=2, figsize=(16, 12))
    
    for x in range(2):
        for y in range(2):
            for signal, label, color in zip(matrices, labels, colors):
                ax[x, y].plot(interval, signal[:, x*2 + y], alpha=0.5, color=color, label=label)
            ax[x, y].set_title(titles[x*2 + y])
            ax[x, y].legend()
    plt.show()

plot_periodical([original, noisy], ['original', 'noisy'], ['blue', 'red'])
```

Next we want to create a mixed signal by applying a mixing matrix to the previously created signal array. To this end you should:

* Standardize the original and the noisy signal arrays using np.std
* Create any invertible $4\times4$ matrix.
* Multiply the previously created original and noisy signal arrays with the mixing matrix and store the resulting arrays.
* Finally plot the standardized, the mixed, the standardized noisy and the mixed noisy signal array. In each plot obviously four different signal graphs should be visible(as you previously were asked to create 4 different signals that should be stored in the array ;))


* <span style="color:rgb(0,120,170)">**TODO:** Code</span>

<h3 style="color:rgb(208,90,80)">Answer 4.7:</h3>

```python
# step 1
def scl(mat):
    return (mat - mat.mean(axis=0)) / np.std(mat, axis=0)

original_scld = scl(original)
noisy_scld = scl(noisy)

# step 2

# quick and dirty way to check if matrix is invertible
# det != 0 is unsafe because of floating point errors
while True:
    mixing_matrix = np.random.randint(-5, 5, (4, 4))
    try: 
        np.linalg.inv(mixing_matrix)
        break
    except np.linalg.LinAlgError:
        print('Created not invertable matrix')

# step 3
original_mxd = original @ mixing_matrix
noisy_mxd = noisy @ mixing_matrix

# step 4
plot_periodical(
    [original_scld, original_mxd, noisy_scld, noisy_mxd], 
    ['original scaled', 'original mixed', 'noisy scaled', 'noisy mixed'],
    ['blue', 'violet', 'red', 'orange'], 
    ['Sine', 'Cosine', 'Tangent', 'Sawtooth']
)
```
